```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Parameters for the binomial distribution 
n = 10 #number trials 
p = 0.5 #probability succesS
#Range of possible successe 
x = np.arange(0, n + 1) 
y = binom.pmf(x, n, p)
plt.figure(figsize=(8, 5)) 
plt.bar(x, y, color="purple", edgecolor="yellow") 
plt.title(f"Binomial Distribution (n={n}, P={p})") 
plt.xlabel('Mumber of Successes') 
plt.ylabel("Probability")
plt.xticks(x) 
plt.grid(axis='y', linestyle='-.', alpha=0.7) 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import bernoulli, binom

# Set up figure for 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Bernoulli and Binomial Distribution Visual Comparison', fontsize=16)

# --- Bernoulli Distribution Plots ---
bernoulli_ps = [0.3, 0.7]
for i, p in enumerate(bernoulli_ps):
    x = [0, 1]
    y = bernoulli.pmf(x, p)
    ax = axes[0, i]
    ax.bar(x, y, width=0.2, color='lightgreen', edgecolor='black')
    ax.set_xticks([0, 1])
    ax.set_ylim(0, 1)
    ax.set_title(f'Bernoulli Distribution (p={p})')
    ax.set_xlabel('Outcome')
    ax.set_ylabel('Probability')
    ax.grid(axis='y', linestyle='--', alpha=0.6)

# --- Binomial Distribution Plots ---
binomial_params = [(10, 0.3), (10, 0.7)]
for i, (n, p) in enumerate(binomial_params):
    x = np.arange(0, n + 1)
    y = binom.pmf(x, n, p)
    ax = axes[1, i]
    mean = np.mean(binomial_params)
    median = np.median(binomial_params)
    #sns.histplot(binomial_params, kde = True, color='skyblue', bins=8)
    plt.axvline(mean,color='red', linestyle = "--", label=f'Mean ={mean:.2f}')
    plt.axvline(median,color='blue', linestyle = "--", label=f'Median ={median:.2f}')
    ax.bar(x, y, color='skyblue', edgecolor='black')
    ax.set_title(f'Binomial Distribution (n={n}, p={p})')
    ax.set_xlabel('Number of Successes')
    ax.set_ylabel('Probability')
    ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
