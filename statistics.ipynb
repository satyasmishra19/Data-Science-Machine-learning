```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
math_scores = [65, 70, 75, 80, 85, 90, 95, 100, 105, 110]
science_scores = [60, 68, 72, 78, 83, 89, 93, 99, 104, 108]
Math = pd.Series(math_scores)
Science = pd.Series(science_scores)
```

```python
mean1 = np.mean(Math)
median1 = np.median(Math)
std_dev1 = np.std(Math)
mean2 = np.mean(Science)
median2 = np.median(Science)
std_dev2 = np.std(Science)
print(mean1)
print(median1)
print(std_dev1)
print(mean2)
print(median2)
print(std_dev2)
```

```python
cov_matrix = np.cov(Math, Science, ddof=0)
covariance = cov_matrix[0,1]
print(f"covariance: {covariance:.2f}")
correlation_matrix = np.corrcoef(Math, Science)
correlation = correlation_matrix[0,1]
print(f"correlation: {correlation:.2f}")
```

```python
" subplot 2 scatter Plot "
plt.subplot(2, 2, 1) 
plt.scatter(Math, Science, color='purple')
#sns.regplot(x=Math, y=Science, color='purple')
plt.title("scatter Plot: Math vs Science") 
plt.xlabel("Math")
plt.ylabel("Science")
plt.grid(True)

" Subplot 3: Heatnap for Correlation Matrix "
plt.subplot(2, 2, 2) 
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", xticklabels=["Math", "Science"], yticklabels=["Math", "Science"]) 
plt.title( 'Correlation Matrix Heatmap')

#plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 3)
sns.boxplot(data=Math,orient='h',color='blue')
plt.title("sns Plot: Math") 
plt.grid(True)

plt.subplot(2, 2, 4)
sns.boxplot(data=Science,orient='h',color='red')
plt.title("sns Plot: Science") 
plt.grid(True)

#Show plots
plt.tight_layout()
plt.show()

```
