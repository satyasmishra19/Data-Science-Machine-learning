```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameters
x = np.arange(0, 10)

# PMF of Poisson distribution
pmf = poisson.pmf(x, mu=10)

# Plot
plt.bar(x, pmf, color='yellow', edgecolor='black')
plt.title("Poisson Distribution (Î» = 3)")
plt.xlabel("Number of events (X)")
plt.ylabel("Probability P(X = x)")
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameters
lambda_hour = 10
lambda_half_hour = 5

# x-axis values for plotting
x_hour = np.arange(0, 21)
x_half_hour = np.arange(0, 11)

# PMF values
pmf_hour = poisson.pmf(x_hour, mu=lambda_hour)
pmf_half_hour = poisson.pmf(x_half_hour, mu=lambda_half_hour)

# a) P(0 customers in 1 hour)
prob_a = poisson.pmf(0, mu=lambda_hour)

# b) P(2 customers in 1 hour)
prob_b = poisson.pmf(2, mu=lambda_hour)

# c) P(2 customers in 30 mins)
prob_c = poisson.pmf(2, mu=lambda_half_hour)

# d) P(<2 customers in 30 mins) = P(0) + P(1)
prob_d = poisson.cdf(1, mu=lambda_half_hour)

# Plot for 1 hour
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(x_hour, pmf_hour, color='skyblue', edgecolor='black')
plt.bar(0, prob_a, color='red', label='P(X=0) = {:.4f}'.format(prob_a))
plt.bar(2, prob_b, color='green', label='P(X=2) = {:.4f}'.format(prob_b))
plt.title("Poisson Distribution (Î» = 10) - 1 Hour")
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Plot for 30 minutes
plt.subplot(1, 2, 2)
plt.bar(x_half_hour, pmf_half_hour, color='lightcoral', edgecolor='black')
plt.bar(2, prob_c, color='blue', label='P(X=2) = {:.4f}'.format(prob_c))
#plt.bar([0,1],prob_f, color='green', label='P(X=2) = {:.4f}'.format(prob_f))
plt.bar([0, 1],poisson.pmf([0, 1], mu=lambda_half_hour), color='orange', label='P(X<2) = {:.4f}'.format(prob_d))
plt.title("Poisson Distribution (Î» = 5) - 30 Minutes")
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Print all probability values
print("Numerical Results:")
print(f"a) P(0 customers in 1 hour): {prob_a:.4f}")
print(f"b) P(2 customers in 1 hour): {prob_b:.4f}")
print(f"c) P(2 customers in 30 mins): {prob_c:.4f}")
print(f"d) P(<2 customers in 30 mins): {prob_d:.4f}")
```
