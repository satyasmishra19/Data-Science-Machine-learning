```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
data = [45,50,52,53,55,60,65,70,75,80,90]
series = pd.Series(data)
```

```python
#Descriptive Statistics
mean = np.mean(series)
median = np.median(series)
mode = series.mode().tolist()
data_range = np.max(series)-np.min(series)
variance = np.var(series)
std_dev = np.std(series)
```

```python
#Quartiles &IQR
q1 = series.quantile(0.25,interpolation='midpoint')
q2 = series.quantile(0.50,interpolation='midpoint')
q3 = series.quantile(0.75,interpolation='midpoint')
iqr = q3-q1
```

```python
print(q1,q2,q3)
```

```python
# OUTLIER dETECTION
l_bound = q1 - 1.5*iqr
u_bound = q3 + 1.5*iqr
outliers = [x for x in series if x < l_bound or x > u_bound]
```

```python
print(l_bound,u_bound)
```

```python
#Skewness and Kurtosis
data_skewness = skew(series)
data_kurtosis = kurtosis(series)
```

```python
print(f"Mean: {mean:.2f},Median: {median}, Mode: {mode}")
print(f"Range: {data_range},Variance: {variance:.2f},Std dev:{std_dev:.2f}")
print(f"Q1:{q1}, Q2:{q2}, Q3:{q3}, IQR: {iqr}")
print(f"Outliers: (outliers if outliers else 'None')")
print(f"skewness: {data_skewness: .2f}")
print(f"Kurtosis: {data_kurtosis: .2f}")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(series, kde = True, color='skyblue', bins=8)
plt.axvline(mean,color='red', linestyle = "--", label=f'Mean ={mean:.2f}')
plt.axvline(median,color='blue', linestyle = "--", label=f'Median ={median:.2f}')
plt.title("skewness")
plt.legend()
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(data=series, orient='h', color='orange')
plt.title( 'Boxplot - Outllers and Spread') 
plt.xlabel('Scores')
plt.grid(True)
plt.show()
```

```python
Company_A = [100, 102, 105, 103, 110, 112, 115, 118, 120, 122] 
Company_B = [80, 82, 85, 87, 88, 92, 95, 97, 98, 100]
```

```python
Company_A =np.array(Company_A)
Company_B =np.array(Company_B)
```

```python
cov_matrix = np.cov(Company_A, Company_B, ddof=0)
covariance = cov_matrix[0,1]
```

```python
correlation_matrix = np.corrcoef(Company_A, Company_B)
correlation = correlation_matrix[0,1]
```

```python
print(f"covariance: {covariance:.2f}")
print(f"correlation: {correlation:.2f}")
```

```python
plt.figure(figsize=(15, 4)) 
# Subplot 1: Line Plot 
plt.subplot(1, 3, 1) 
plt.plot(Company_A, label='Company A', marker='o') 
plt.plot(Company_B, label='Company B', marker='s') 
plt.title('stock Prices Over Tino')
plt.xlabel("Time (Days)") 
plt.ylabel('price') 
plt.legend()
plt.grid(True)

" subplot 2 scatter Plot "
plt.subplot(1, 3, 2) 
plt.scatter(Company_A, Company_B, color='purple')
plt.title("scatter Plot: Company A vs Coxpany B") 
plt.xlabel("Company A Price")
plt.ylabel("Company B Price")
plt.grid(True)

" Subplot 3: Heatnap for Correlation Matrix "
plt.subplot(1, 3, 3) 
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", xticklabels=["A", "B"], yticklabels=["A", "B"]) 
plt.title( 'Correlation Matrix Heatmap')
#Show plots 
plt.tight_layout() 
plt.show()
```
